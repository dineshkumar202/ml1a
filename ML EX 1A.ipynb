8. Display the row index along with outlier values and Z-score for the column Guarantee_Period and verify with box plot (Python code)

```python
print("URK22CS1173")
df = pd.DataFrame(data)

df['Z_score'] = (df['Guarantee_Period'] - df['Guarantee_Period'].mean()) / df['Guarantee_Period'].std()

outliers = df[(df['Z_score'] > 3) | (df['Z_score'] < -3)]

print("Outliers in 'Guarantee_Period':")
print(outliers[['Guarantee_Period', 'Z_score']])

plt.figure(figsize=(8, 6))
plt.boxplot(df['Guarantee_Period'], vert=False)
plt.title('Box Plot of Guarantee_Period')
plt.xlabel('Guarantee_Period')
plt.show()
```

12. Analyze the skewness of the feature using plot distribution graph for the cc column
and display whether the feature is right skew, left skew or no skew (Python code)

```python
print("URK22CS1173")
from scipy.stats import skew
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df['cc'], kde=True, color='blue') 
plt.title("Distribution of 'cc' Column")
plt.xlabel("cc")
plt.ylabel("Frequency")
plt.show()

cc_skewness = skew(df['cc'], nan_policy='omit')

if cc_skewness > 0:
    skew_type = 'Right Skew (Positively Skewed)'
elif cc_skewness < 0:
    skew_type = 'Left Skew (Negatively Skewed)'
else:
    skew_type = 'No Skew (Symmetrical Distribution)'

print(f"Skewness of 'cc' column: {cc_skewness:.2f}")
print(f"Interpretation: {skew_type}")
```
